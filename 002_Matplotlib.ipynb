```python
from matplotlib import pyplot as plt
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300, 500, 850, 990, 1378.7, 1500.9, 1753.9]

plt.plot(years, gdp, color='blue', marker='o')
plt.title("Nominal GDP over the years 1950-2010")
plt.ylabel("Billions of USD")
plt.xlabel("Years")
plt.show()
```
